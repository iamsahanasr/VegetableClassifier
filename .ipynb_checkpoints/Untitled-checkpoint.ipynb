```python
import os
vegetables = os.listdir(base_dir)
features = []
labels = []
base_dir = os.path.join(os.getcwd(),"dataset")
for vegetable in os.listdir(base_dir):
    for image in os.listdir(os.path.join(base_dir,vegetable)):
        im = io.imread(os.path.join(base_dir,vegetable,image))
        im = resize(im,(100,100)).flatten()
        features.append(im)
        labels.append(vegetables.index(vegetable))
        
print(features)
print(labels)
        
        
    
```

```python
def euclidean(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))

print(euclidean(features[0],features[3]))
```

```python
test_image = io.imread("test_cucumber.jpg")
test_image = resize(test_image,(100,100)).flatten()
distances = np.array([euclidean(test_image,image) for image in features])
prediction = distances.argmin()
# print(vegetables)
print(vegetables[labels[prediction]])
```
